```python
import requests
import pandas as pd

GITHUB_TOKEN = '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'
HEADERS = {'Authorization': f'token {GITHUB_TOKEN}'}

def fetch_users(location='London', min_followers=500):
    users = []
    page=1

    while page!=5:#check for how to prpocesss more pages or more data
        url = f"https://api.github.com/search/users?q=location:{location}+followers:>{min_followers}&per_page=100&page={page}"
        response = requests.get(url, headers=HEADERS).json()
        users.extend(response['items'])
        page+=1
        #####code for debugging#####
        #print(page)
        #time.sleep(2)
        ###########################
    return users

def fetch_repositories(user_login):
    repos = []
    page=1
    # url = f"https://api.github.com/users/{user_login}/repos?per_page=100"
    #while page!=2:
    url = f"https://api.github.com/users/{user_login}/repos?per_page=10"
    response = requests.get(url, headers=HEADERS).json()
    repos.extend(response)
    return repos

users = fetch_users()
user_data = []
repo_data = []

def is_null(val):
    if val is None:
        return ''
    else:
        val.strip().lstrip('@').upper()

for user in users:
    user_detail = requests.get(user['url'], headers=HEADERS).json()
    user_data.append({
        'login': user_detail['login'],
        'name': user_detail.get('name', ''),
        'company': user_detail.get('company', ''),
        'location': user_detail.get('location', ''),
        'email': user_detail.get('email', ''),
        'hireable': user_detail.get('hireable', ''),
        'bio': user_detail.get('bio', ''),
        'public_repos': user_detail.get('public_repos', 0),
        'followers': user_detail.get('followers', 0),
        'following': user_detail.get('following', 0),
        'created_at': user_detail.get('created_at', '')
    })

    repos = fetch_repositories(user_detail['login'])[:100]
    for repo in repos:
        repo_data.append({
            'login': user_detail['login'],
            'full_name': repo['full_name'],
            'created_at': repo['created_at'],
            'stargazers_count': repo['stargazers_count'],
            'watchers_count': repo['watchers_count'],
            'language': repo.get('language', ''),
            'has_projects': repo.get('has_projects', False),
            'has_wiki': repo.get('has_wiki', False),
            'license_name': repo.get('license')
        })

# Save to CSV
users_df = pd.DataFrame(user_data)
users_df.to_csv('users.csv', index=False)

repos_df = pd.DataFrame(repo_data)
repos_df.to_csv('repositories.csv', index=False)
```

####4. Company with the Most Developers:


```python
import pandas as pd

# Example: Load your DataFrame (replace this with your actual loading method)
# repos_df = pd.read_csv('your_repos_file.csv')

# Print the first few rows of the DataFrame to inspect the data
print("Columns in DataFrame:", repos_df.columns)
print("First few rows of the DataFrame:")
print(repos_df.head())

# Check if the 'company' column exists
if 'company' not in repos_df.columns:
    raise ValueError("The DataFrame must contain a 'company' column.")
# Drop rows where 'company' is null or empty
repos_df = repos_df[repos_df['company'].notna() & (repos_df['company'] != '')]

# Print the number of rows after dropping nulls
print(f"Number of rows after dropping null or empty company values: {repos_df.shape[0]}")

# Clean up company names
repos_df['company'] = repos_df['company'].str.strip()  # Trim whitespace
repos_df['company'] = repos_df['company'].str.lstrip('@')  # Strip leading '@'
repos_df['company'] = repos_df['company'].str.upper()  # Convert to uppercase

# Check the cleaned company names
print("Cleaned company names:")
print(repos_df['company'].unique())
# Count occurrences of each company
company_counts = repos_df['company'].value_counts()

# Print the company counts
print("Company counts:")
print(company_counts)

# Identify the most common company
if not company_counts.empty:
    most_common_company = company_counts.idxmax()  # Get the name of the most common company
    most_common_count = company_counts.max()  # Get the count of the most common company

    # Print the result
    print(f"Most common company: {most_common_company} with {most_common_count} users")
else:
    print("No companies found in the dataset.")

print(repos_df.dtypes)
```

########11. Correlation Between Projects and Wiki Enabled:


```python
import pandas as pd

# Step 1: Load your DataFrame (replace with your actual loading method)
# repos_df = pd.read_csv('your_repos_file.csv')

# Step 2: Check if 'has_projects' and 'has_wiki' columns exist
if 'has_projects' not in repos_df.columns or 'has_wiki' not in repos_df.columns:
    raise ValueError("The DataFrame must contain 'has_projects' and 'has_wiki' columns.")

# Step 3: Print the first few rows of relevant columns for inspection
print("First few rows of the relevant columns:")
print(repos_df[['has_projects', 'has_wiki']].head())

# Step 4: Check for unique values in 'has_projects' and 'has_wiki'
print("Unique values for 'has_projects':", repos_df['has_projects'].unique())
print("Unique values for 'has_wiki':", repos_df['has_wiki'].unique())

# Step 5: Convert boolean columns to integers (1 for True, 0 for False)
repos_df['has_projects'] = repos_df['has_projects'].astype(int)
repos_df['has_wiki'] = repos_df['has_wiki'].astype(int)

# Step 6: Check for any missing values
if repos_df['has_projects'].isnull().any() or repos_df['has_wiki'].isnull().any():
    print("Warning: Missing values found in 'has_projects' or 'has_wiki'. Dropping missing values.")
    repos_df = repos_df.dropna(subset=['has_projects', 'has_wiki'])

# Step 7: Calculate the correlation
correlation = repos_df['has_projects'].corr(repos_df['has_wiki'])

# Step 8: Print the correlation rounded to 3 decimal places
print(f"Correlation between having projects enabled and having wiki enabled: {correlation:.3f}")
```

########12. Hireable Users Following More People:


```python
import pandas as pd

# Load your DataFrame (replace with your actual loading method)
# users_df = pd.read_csv('your_users_file.csv')

# Step 1: Check if the DataFrame contains the necessary columns
print("DataFrame columns:", users_df.columns)
print("First few rows of the DataFrame:")
print(users_df.head())

# Step 2: Calculate average following for hireable users
hireable_avg_following = users_df[users_df['hireable'] == True]['following'].mean()

# Step 3: Calculate average following for non-hireable users
non_hireable_avg_following = users_df[users_df['hireable'] == False]['following'].mean()

# Step 4: Calculate the difference
average_difference = hireable_avg_following - non_hireable_avg_following

# Step 5: Print the result rounded to 3 decimal places
print(f"Average following difference (hireable - non-hireable): {average_difference:.3f}")
```

########13. Regression Slope: Followers vs. Bio Word Count:


```python
from scipy.stats import linregress

# Filter out users without a bio
users_with_bio = users_df.dropna(subset=['bio'])

# Calculate the word count of each bio (split by whitespace)
users_with_bio['bio_word_count'] = users_with_bio['bio'].apply(lambda x: len(x.split()))

# Perform linear regression with bio word count as the predictor and followers as the outcome
slope, intercept, r_value, p_value, std_err = linregress(users_with_bio['bio_word_count'], users_with_bio['followers'])

# Print the slope to 3 decimal places
print(f"Regression slope of followers on bio word count: {slope:.3f}")
```

########14. Top 5 Users by Weekend Repository Creation:


```python
import pandas as pd

# Load your DataFrame (replace with your actual loading method)
# repos_df = pd.read_csv('your_repos_file.csv')

# Step 1: Check if DataFrame loaded correctly and contains required columns
print("DataFrame columns:", repos_df.columns)
print("First few rows of the DataFrame:")
print(repos_df.head())

# Step 2: Convert 'created_at' to datetime if it's not already
repos_df['created_at'] = pd.to_datetime(repos_df['created_at'], errors='coerce')

# Step 3: Check for any NaT values after conversion
if repos_df['created_at'].isnull().any():
    print("Warning: There are NaT values in 'created_at'. Dropping these rows.")
    repos_df = repos_df.dropna(subset=['created_at'])

# Step 4: Identify weekends (5 = Saturday, 6 = Sunday)
repos_df['is_weekend'] = repos_df['created_at'].dt.dayofweek >= 5

# Step 5: Filter for weekend repositories
weekend_repos = repos_df[repos_df['is_weekend']]

# Step 6: Check the number of weekend repositories
print(f"Total weekend repositories: {len(weekend_repos)}")

# Step 7: Group by 'login' and count the number of repositories created
top_weekend_creators = weekend_repos.groupby('login').size().nlargest(5).index.tolist()

# Step 8: Print the top 5 users' logins in order, comma-separated
print("Top 5 users who created the most repositories on weekends:")
print(','.join(top_weekend_creators))

```

########15. Do Hireable Users Share Emails More Often?


```python
import pandas as pd

# Sample Data: Replace this with your actual DataFrame loading
data = {
    'login': ['user1', 'user2', 'user3', 'user4', 'user5'],
    'email': ['user1@example.com', None, 'user3@example.com', 'user4@example.com', None],
    'hireable': ['true', 'false', 'true', 'true', 'false'],
}


users_df = pd.DataFrame(data)

# Step 1: Clean the 'hireable' column and convert it to boolean
if users_df['hireable'].dtype == 'object':
    users_df['hireable'] = users_df['hireable'].str.lower().map({'true': True, 'false': False})

# Step 2: Count total users in each category (hireable and non-hireable)
total_hireable = users_df[users_df['hireable']].shape[0]
total_non_hireable = users_df[~users_df['hireable']].shape[0]

# Step 3: Calculate the number of users with email in each category
hireable_with_email = users_df[users_df['hireable']]['email'].notnull().sum()
non_hireable_with_email = users_df[~users_df['hireable']]['email'].notnull().sum()

# Step 4: Calculate the fractions
fraction_hireable_with_email = hireable_with_email / total_hireable if total_hireable > 0 else 0
fraction_non_hireable_with_email = non_hireable_with_email / total_non_hireable if total_non_hireable > 0 else 0

# Step 5: Calculate the difference
email_difference = fraction_hireable_with_email - fraction_non_hireable_with_email

# Step 6: Print the result rounded to 3 decimal places
print(f"Fraction difference (hireable - non-hireable): {email_difference:.3f}")

# data = {
#     'login': ['user1', 'user2', 'user3', 'user4', 'user5'],
#     'email': ['user1@example.com', None, 'user3@example.com', 'user4@example.com', None],
#     'hireable': ['true', 'false', 'true', 'true', 'false'],
# }
```
